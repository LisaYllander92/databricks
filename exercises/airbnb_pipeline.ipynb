```python
import dlt # imports Databricks DLT-library  

# tells that it's a DLT-table with name and metadata
@dlt.table(
    name="bronze_airbnb",
    comment="Raw Airbnb data from csv file",
    table_properties = {"quality": "bronze"}
)

def bronze_airbnb():
    return (
        spark.readStream # makes it into a streaming table
        .format("cloudFiles") #Databricks way to stream files from Volumes/storage
        .option("cloudFiles.format", "csv") # what type of file the data is stored in
        .option("header", "true")
        .load("/Volumes/airbnb/hosts/csv_files") # point to where the data is stored (the folder, NOT the file)
    )
```
